```python
import os
import math
import random
import shutil
import zipfile
from pathlib import Path
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader

from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error

import optuna
import statsmodels.api as sm
import joblib

# Seed for reproducibility
RANDOM_SEED = 42
torch.manual_seed(RANDOM_SEED)
np.random.seed(RANDOM_SEED)
random.seed(RANDOM_SEED)

# Project folders
os.makedirs("output", exist_ok=True)     # store model, scaler, plots
os.makedirs("output/repo", exist_ok=True) # GitHub-ready repo later

print("Folders created: output/, output/repo/")
```

```python
def make_synthetic_series(n=3000, noise_std=0.3):
    t = np.arange(n)
    trend = 0.0005 * (t ** 1.5)
    daily = 2 * np.sin(2 * np.pi * t / 24)
    weekly = 1.5 * np.sin(2 * np.pi * t / (24 * 7))
    noise = np.random.normal(scale=noise_std, size=n)
    base = 10 + trend + daily + weekly + noise
    idx = pd.date_range("2000-01-01", periods=n, freq="H")
    return pd.Series(base, index=idx, name="y")

series = make_synthetic_series()
series.plot(figsize=(12,4), title="Synthetic Multi-Seasonal Series")
plt.show()

len(series)
```

```python
def create_windows(series, input_len, output_len):
    arr = series.values
    windows = []
    for i in range(len(arr) - input_len - output_len):
        x = arr[i:i+input_len]
        y = arr[i+input_len:i+input_len+output_len]
        windows.append((x.astype(np.float32), y.astype(np.float32)))
    return windows

class TimeSeriesDataset(Dataset):
    def __init__(self, windows, scaler=None):
        self.windows = windows
        self.scaler = scaler

    def __len__(self):
        return len(self.windows)

    def __getitem__(self, idx):
        x, y = self.windows[idx]
        if self.scaler:
            x = self.scaler.transform(x.reshape(-1,1)).reshape(-1)
            y = self.scaler.transform(y.reshape(-1,1)).reshape(-1)
        return torch.tensor(x).float().unsqueeze(0), torch.tensor(y).float()

print("Window & Dataset utilities ready!")
```

```python
class TrendBasis(nn.Module):
    def __init__(self, backcast_length, forecast_length, degree=2):
        super().__init__()
        b = np.linspace(0, 1, backcast_length)
        f = np.linspace(0, 1, forecast_length)
        self.register_buffer("bt", torch.tensor(b).float().unsqueeze(0))
        self.register_buffer("ft", torch.tensor(f).float().unsqueeze(0))
        self.degree = degree
        self.forecast_length = forecast_length # Store forecast_length

    def forward(self, theta):
        d = self.degree + 1
        bc = theta[:, :d]
        fc = theta[:, d:2*d]
        B = torch.stack([self.bt**i for i in range(d)], dim=1)
        F = torch.stack([self.ft**i for i in range(d)], dim=1)
        back = torch.einsum("bd,bdl->bl", bc, B.repeat(theta.size(0),1,1))
        fore = torch.einsum("bd,bdl->bl", fc, F.repeat(theta.size(0),1,1))
        return back, fore

class SeasonalityBasis(nn.Module):
    def __init__(self, backcast_length, forecast_length, harmonics=10):
        super().__init__()
        b = np.linspace(0, 1, backcast_length)
        f = np.linspace(0, 1, forecast_length)
        self.register_buffer("bt", torch.tensor(b).float().unsqueeze(0))
        self.register_buffer("ft", torch.tensor(f).float().unsqueeze(0))
        self.h = harmonics
        self.forecast_length = forecast_length # Store forecast_length

    def forward(self, theta):
        h = self.h
        bc = theta[:, :2*h]
        fc = theta[:, 2*h:4*h]

        B = torch.stack([torch.cos(2*math.pi*(i+1)*self.bt) for i in range(h)] +
                        [torch.sin(2*math.pi*(i+1)*self.bt) for i in range(h)], dim=1)

        F = torch.stack([torch.cos(2*math.pi*(i+1)*self.ft) for i in range(h)] +
                        [torch.sin(2*math.pi*(i+1)*self.ft) for i in range(h)], dim=1)

        back = torch.einsum("bd,bdl->bl", bc, B.repeat(theta.size(0),1,1))
        fore = torch.einsum("bd,bdl->bl", fc, F.repeat(theta.size(0),1,1))
        return back, fore

class NBeatsBlock(nn.Module):
    def __init__(self, input_len, output_len, block_type="generic", hidden=128):
        super().__init__()
        self.block_type = block_type

        self.fc = nn.Sequential(
            nn.Linear(input_len, hidden), nn.ReLU(),
            nn.Linear(hidden, hidden), nn.ReLU(),
        )

        if block_type == "trend":
            self.basis = TrendBasis(input_len, output_len, degree=2)
            theta_size = 6
        elif block_type == "season":
            self.basis = SeasonalityBasis(input_len, output_len, harmonics=8)
            theta_size = 32
        else:
            self.basis = None
            theta_size = input_len + output_len

        self.theta = nn.Linear(hidden, theta_size)

    def forward(self, x):
        x = x.squeeze(1)
        t = self.theta(self.fc(x))

        if self.basis:
            back, fore = self.basis(t)
        else:
            back = t[:, :x.shape[1]]
            fore = t[:, x.shape[1]:]
        return back, fore

class NBeatsModel(nn.Module):
    def __init__(self, input_len, output_len):
        super().__init__()
        self.blocks = nn.ModuleList([
            NBeatsBlock(input_len, output_len, "trend"),
            NBeatsBlock(input_len, output_len, "season"),
            NBeatsBlock(input_len, output_len, "generic"),
        ])

    def forward(self, x):
        x = x.squeeze(1)
        res = x
        # Removed the problematic line that caused negative dimension error
        # forecast = torch.zeros(x.size(0), self.blocks[0].theta.out_features - x.size(1))

        forecast = torch.zeros((x.size(0), self.blocks[0].basis.forecast_length), device=x.device)

        for b in self.blocks:
            back, fore = b(res)
            res = res - back
            forecast = forecast + fore
        return forecast
```

```python
INPUT_LEN = 64
FORECAST_LEN = 24

n = len(series)
train = series[:int(0.7*n)]
val   = series[int(0.7*n):int(0.85*n)]
test  = series[int(0.85*n):]

scaler = StandardScaler()
scaler.fit(train.values.reshape(-1,1))

train_ws = create_windows(train, INPUT_LEN, FORECAST_LEN)
val_ws = create_windows(val, INPUT_LEN, FORECAST_LEN)
test_ws = create_windows(test, INPUT_LEN, FORECAST_LEN)

train_ds = TimeSeriesDataset(train_ws, scaler)
val_ds = TimeSeriesDataset(val_ws, scaler)
test_ds = TimeSeriesDataset(test_ws, scaler)

train_loader = DataLoader(train_ds, batch_size=64, shuffle=True)
val_loader   = DataLoader(val_ds, batch_size=64)
test_loader  = DataLoader(test_ds, batch_size=64)

print("Train windows:", len(train_ws))
print("Val windows:", len(val_ws))
print("Test windows:", len(test_ws))
```

```python
def train(model, train_loader, val_loader, epochs=10, lr=1e-3):
    device = "cuda" if torch.cuda.is_available() else "cpu"
    model.to(device)
    opt = torch.optim.Adam(model.parameters(), lr=lr)
    loss_fn = nn.MSELoss()

    for ep in range(1, epochs+1):
        model.train()
        tr_losses = []

        for x, y in train_loader:
            x, y = x.to(device), y.to(device)
            opt.zero_grad()
            pred = model(x)
            loss = loss_fn(pred, y)
            loss.backward()
            opt.step()
            tr_losses.append(loss.item())

        model.eval()
        val_losses = []
        with torch.no_grad():
            for x, y in val_loader:
                x, y = x.to(device), y.to(device)
                pred = model(x)
                val_losses.append(loss_fn(pred, y).item())

        print(f"Epoch {ep}: train={np.mean(tr_losses):.4f}, val={np.mean(val_losses):.4f}")
```

```python
model = NBeatsModel(INPUT_LEN, FORECAST_LEN)
train(model, train_loader, val_loader, epochs=8, lr=1e-3)
print("Training complete.")
```

```python
device = "cuda" if torch.cuda.is_available() else "cpu"
model.to(device)
model.eval()

preds, trues = [], []

with torch.no_grad():
    for x, y in test_loader:
        x = x.to(device)
        p = model(x).cpu().numpy()
        preds.append(p)
        trues.append(y.numpy())

preds = np.vstack(preds)
trues = np.vstack(trues)

# inverse-scale
preds_inv = scaler.inverse_transform(preds.reshape(-1,1)).reshape(preds.shape)
trues_inv = scaler.inverse_transform(trues.reshape(-1,1)).reshape(trues.shape)

def rmse(y_true, y_pred):
    return np.sqrt(mean_squared_error(y_true, y_pred))

def smape(y_true, y_pred):
    numerator = np.abs(y_pred - y_true)
    denominator = (np.abs(y_true) + np.abs(y_pred)) / 2
    return np.mean(numerator / denominator) * 100

print("MAE :", mean_absolute_error(trues_inv.flatten(), preds_inv.flatten()))
print("RMSE:", rmse(trues_inv.flatten(), preds_inv.flatten()))
print("sMAPE:", smape(trues_inv.flatten(), preds_inv.flatten()))

plt.plot(trues_inv[0], label="true")
plt.plot(preds_inv[0], label="pred")
plt.title("Forecast Example")
plt.legend()
plt.show()
```

```python
from statsmodels.tsa.stattools import adfuller
from pmdarima.arima.utils import ndiffs, nsdiffs

# 1. Confirm data splits (already done in a previous cell, just re-referencing)
print(f"Train series length: {len(train)}")
print(f"Validation series length: {len(val)}")
print(f"Test series length: {len(test)}")

# 2. Use Augmented Dickey-Fuller (ADF) test to check stationarity of train dataset
print("\nPerforming Augmented Dickey-Fuller test on train dataset...")
adf_test = adfuller(train)
print(f"ADF Statistic: {adf_test[0]:.4f}")
print(f"P-value: {adf_test[1]:.4f}")
print("Is the series stationary? (P-value <= 0.05):", adf_test[1] <= 0.05)

# 3. Estimate optimal non-seasonal differencing order (d)
print("\nEstimating optimal non-seasonal differencing order (d)... ")
# Use 'test'='adf' (Augmented Dickey-Fuller) or 'kpss' (Kwiatkowski-Phillips-Schmidt-Shin)
# or 'ch' (Chow-Liu) for stationarity test.
d_order = ndiffs(train, test='adf')
print(f"Estimated non-seasonal differencing order (d): {d_order}")

# 4. Estimate optimal seasonal differencing order (D)
print("\nEstimating optimal seasonal differencing order (D) with m=24...")
m = 24 # Hourly data, daily seasonality
D_order = nsdiffs(train, m=m, test='ocsb') # OCSB test is generally recommended for seasonal differencing
print(f"Estimated seasonal differencing order (D): {D_order}")
```

## Define Optuna Objective Function for ARIMA




```python
from pmdarima.arima import ARIMA
from sklearn.metrics import mean_squared_error

def objective(trial):
    # 3. Suggest Hyperparameters
    p = trial.suggest_int('p', 0, 3)
    q = trial.suggest_int('q', 0, 3)
    P = trial.suggest_int('P', 0, 2) # P typically lower than p
    Q = trial.suggest_int('Q', 0, 2) # Q typically lower than q

    # 4. Fix d parameter (non-seasonal differencing order)
    d = d_order # Use the previously estimated d_order

    # 5. Fix D parameter (seasonal differencing order)
    D = D_order # Use the previously estimated D_order

    # 6. Set seasonal period
    m = 24 # Hourly data, daily seasonality

    # 7. Instantiate and Train ARIMA
    try:
        model = ARIMA(
            order=(p, d, q),
            seasonal_order=(P, D, Q, m),
            suppress_warnings=True,
            stepwise=False, # Disable stepwise to ensure full search with Optuna
            n_jobs=-1 # Use all available cores
        )
        model.fit(train)

        # 9. Generate predictions on the val dataset
        predictions = model.predict(n_periods=len(val))

        # 10. Calculate RMSE
        rmse = np.sqrt(mean_squared_error(val, predictions))

    except Exception as e:
        # If the model fails to fit or predict (e.g., singular matrix),
        # return a very high RMSE to discourage these parameters.
        # print(f"Trial failed with parameters p={p}, d={d}, q={q}, P={P}, D={D}, Q={Q}, m={m}: {e}")
        rmse = float('inf')

    # 11. Return the calculated RMSE
    return rmse

print("Optuna objective function for ARIMA is defined.")
```
